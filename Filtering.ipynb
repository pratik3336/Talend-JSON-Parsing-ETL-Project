```python


def read_jsons_to_dataframe(path):
    dfs = []  # List to hold DataFrames
    
    # Use glob to find all JSON files in the specified path, including subdirectories
    json_files = glob.glob(path)
    
    for file in json_files:
        with open(file, 'r') as f:
            # Load the JSON data
            data = json.load(f)
            
            # If the data is a dictionary, wrap it in a list to ensure consistency
            if isinstance(data, dict):
                data = [data]
            
            # Create a DataFrame, potentially flattening the JSON structure
            df = pd.json_normalize(data)
            
            # Append the DataFrame to the list
            dfs.append(df)
    
    # Concatenate all DataFrames, if any
    if dfs:
        final_df = pd.concat(dfs, ignore_index=True)
    else:
        final_df = pd.DataFrame()
    
    return final_df

```
